```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```python
df = spark.read.csv('appl_stock.csv', inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.head(3)
```

```python
df.head(1)[0]
```

```python
# With the filter function, we can add a condition into a specific column.
# in this case we are filtering all rows that have a Close value less than 500.
df.filter("Close < 500").show()
```

```python
# We can show only one column or multiple with the select function.
df.filter("Close < 500").select(['Open','Close']).show()
```

```python
# Instead of using the SQL format, we will use the pyspark syntax
df.filter(df['Close'] < 500).select('Volume').show()
```

```python
# Example of filtering based on multiple conditions.
# Imagine that we want the result of the filter when we have a close price less than 500 and open price greater than 200

# Be carefull to use parenthesis if we are using ampersand or pipe operators

df.filter((df['Close'] < 500) & (df['Open'] > 200)).show()
```

```python
# What date was the low price $197 and 16 cents
df.filter(df['Low']==197.16).select('Date').show()
```

```python
# Instead of using .show we can use .collect and then we will get a list of the actuall row object which means
# I can then save this as a result
results = df.filter(df['Low']==197.16).select('Date').collect()
```
